# Inisialisasi Faiss Retriever

```python
# ==========================================
# CELL 2: CLASS FAISS RETRIEVER
# ==========================================
import faiss
import numpy as np
import pandas as pd
import pyterrier as pt
from sentence_transformers import SentenceTransformer

class FaissRetriever(pt.Transformer):
    def __init__(self, model_name, index_path=None, doc_df=None, batch_size=32):
        self.model = SentenceTransformer(model_name)
        
        # Patch untuk mencegah error parameter tokenizer
        if hasattr(self.model, 'tokenizer') and 'token_type_ids' in self.model.tokenizer.model_input_names:
            self.model.tokenizer.model_input_names.remove('token_type_ids')
            
        self.index = None
        self.doc_map = {} 
        self.batch_size = batch_size
        
        if doc_df is not None:
            self._index_docs(doc_df)
            
    def _index_docs(self, doc_df):
        print(f"Encoding {len(doc_df)} dokumen ke vektor (bisa agak lama)...")
        # PENTING: Untuk Skenario 3, pastikan kolom yang masuk adalah teks Arab yang sudah di-preprocess
        docs = doc_df['text_processed'].tolist()
        docnos = doc_df['docno'].tolist()
        
        embeddings = self.model.encode(docs, batch_size=self.batch_size, show_progress_bar=True, convert_to_numpy=True)
        faiss.normalize_L2(embeddings)
        
        dim = embeddings.shape[1]
        self.index = faiss.IndexFlatIP(dim)
        self.index.add(embeddings)
        
        self.doc_map = {i: docno for i, docno in enumerate(docnos)}
        print("✅ Indexing selesai.")

    def transform(self, queries):
        q_texts = queries['query'].tolist()
        q_embeddings = self.model.encode(q_texts, batch_size=self.batch_size, show_progress_bar=False, convert_to_numpy=True)
        faiss.normalize_L2(q_embeddings)
        
        k = 1000
        scores, ids = self.index.search(q_embeddings, k)
        
        results = []
        for i, qid in enumerate(queries['qid']):
            for j in range(k):
                idx = ids[i][j]
                if idx != -1:
                    results.append({
                        'qid': str(qid),
                        'docno': self.doc_map[idx],
                        'score': scores[i][j],
                        'rank': j + 1
                    })
        return pd.DataFrame(results)
```

# Loading Data 

```python
# ==========================================
# CELL 3: PERSIAPAN DATA (DOKUMEN & KUERI ARAB)
# ==========================================
from arabic_preprocessing import preprocess_pipeline

DATA_PATH = './skripsi-clir-code/data'
RAW_DOCS = os.path.join(DATA_PATH, 'raw/fathul_muin.csv')
QUERY_ARAB = os.path.join(DATA_PATH, 'queries/Kueri_Assesor_Zona2 - queries_arab.csv')
QRELS_PATH = os.path.join(DATA_PATH, 'queries/Kueri_Assesor_Zona2 - qrels.csv')

print("1. Memuat & Preprocessing Dokumen...")
df_docs = pd.read_csv(RAW_DOCS)
df_docs['text_processed'] = df_docs['text'].apply(preprocess_pipeline)
df_docs['docno'] = df_docs['docno'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)

print("2. Memuat Kueri Bahasa Arab (Translate-then-Retrieve)...")
df_queries_arab = pd.read_csv(QUERY_ARAB)
topics = df_queries_arab[['qid', 'query_arab']].rename(columns={'query_arab': 'query'})
topics['qid'] = topics['qid'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)

print("3. Memuat Qrels...")
qrels = pd.read_csv(QRELS_PATH)
# Fix nama kolom jika diperlukan
if 'doc_no' in qrels.columns:
    qrels.rename(columns={'doc_no': 'docno'}, inplace=True)
elif 'doc_id' in qrels.columns:
    qrels.rename(columns={'doc_id': 'docno'}, inplace=True)
    
qrels['qid'] = qrels['qid'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)
qrels['docno'] = qrels['docno'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)
qrels['label'] = qrels['label'].astype(int)

print("✅ Persiapan Data Selesai!")
```

# Evaluasi

```python
# ==========================================
# CELL 4: EVALUASI SKENARIO 3 (AraDPR)
# ==========================================
print("Membangun Index Dense menggunakan model Monolingual: abdoelsayed/AraDPR...")
ARADPR_MODEL_NAME = 'abdoelsayed/AraDPR'

# Gunakan batch_size 16 untuk mencegah CUDA Out of Memory
retriever_aradpr = FaissRetriever(
    model_name=ARADPR_MODEL_NAME, 
    doc_df=df_docs[['docno', 'text_processed']], 
    batch_size=16 
)

print("\nMenjalankan Eksperimen PyTerrier...")
hasil_skenario_3 = pt.Experiment(
    [retriever_aradpr],
    topics, 
    qrels,
    eval_metrics=["recip_rank", "success_10", "success_20", "recall_10", "recall_20"],
    names=["Skenario 3 (AraDPR)"],
    validate="ignore",
    # perquery=True
)

print("\n=== HASIL EVALUASI SKENARIO 3 ===")
display(hasil_skenario_3)

hasil_skenario_3.to_csv('hasil_skenario_3.csv', index=False)
```

```python
# ==========================================
# CELL: EVALUASI OVERALL & PER-TIPE KUERI (FIXED)
# ==========================================
import pyterrier as pt
import pandas as pd
import os
from IPython.display import display

print("1. Menjalankan Evaluasi Per-Kueri (perquery=True)...")
hasil_eval_perq = pt.Experiment(
    [retriever_aradpr], # Ganti dengan retriever yang sesuai di notebook lain
    topics,             # Gunakan topics (Skenario 1/3) atau df_queries (Skenario 2)
    qrels,
    eval_metrics=["recip_rank", "success_10", "success_20"],
    names=["AraDPR Model"], 
    validate="ignore",
    perquery=True 
)

# [PERBAIKAN KUNCI] Mengubah format panjang (measure, value) menjadi kolom metrik biasa
hasil_pivot = hasil_eval_perq.pivot_table(
    index=['name', 'qid'],
    columns='measure',
    values='value'
).reset_index()

print("2. Memuat Metadata Tipe Kueri dan Menggabungkannya...")
queries_indo_path = './skripsi-clir-code/data/queries/Kueri_Assesor_Zona2 - queries_indo.csv'
df_queries = pd.read_csv(queries_indo_path)

# Pastikan tipe data qid sama sebelum di-merge (keduanya harus string)
hasil_pivot['qid'] = hasil_pivot['qid'].astype(str)
df_queries['qid'] = df_queries['qid'].astype(str)

# Kita ambil kolom qid dan query_type dari dataframe kueri
hasil_dengan_tipe = pd.merge(
    hasil_pivot,
    df_queries[['qid', 'query_type']],
    on='qid',
    how='left'
)

# ==========================================
# 3. MENAMPILKAN HASIL KESELURUHAN (OVERALL)
# ==========================================
print("\n" + "="*50)
print("🎯 HASIL KESELURUHAN (ALL QUERIES)")
print("="*50)
overall_metrics = hasil_dengan_tipe.groupby('name')[['recip_rank', 'success_10', 'success_20']].mean().reset_index()

# Ubah format menjadi persentase untuk Success (Hits) agar mudah dibaca di skripsi
overall_metrics['success_10 (%)'] = (overall_metrics['success_10'] * 100).round(2)
overall_metrics['success_20 (%)'] = (overall_metrics['success_20'] * 100).round(2)
overall_metrics['recip_rank'] = overall_metrics['recip_rank'].round(4)

display(overall_metrics[['name', 'recip_rank', 'success_10 (%)', 'success_20 (%)']])

# ==========================================
# 4. MENAMPILKAN HASIL BERDASARKAN TIPE KUERI
# ==========================================
print("\n" + "="*50)
print("📊 HASIL BERDASARKAN TIPE KUERI")
print("="*50)

# Groupby berdasarkan Nama Model dan Tipe Kueri
per_type_metrics = hasil_dengan_tipe.groupby(['name', 'query_type'])[['recip_rank', 'success_10', 'success_20']].mean().reset_index()

per_type_metrics['success_10 (%)'] = (per_type_metrics['success_10'] * 100).round(2)
per_type_metrics['success_20 (%)'] = (per_type_metrics['success_20'] * 100).round(2)
per_type_metrics['recip_rank'] = per_type_metrics['recip_rank'].round(4)

display(per_type_metrics[['name', 'query_type', 'recip_rank', 'success_10 (%)', 'success_20 (%)']])

# Opsional: Simpan ke CSV untuk rekap laporan skripsi
nama_file_output = 'evaluasi_tipe_kueri_skenario_3.csv'
per_type_metrics.to_csv(nama_file_output, index=False)
print(f"\n📊 Tabel analisis berhasil disimpan ke '{nama_file_output}'")
```
