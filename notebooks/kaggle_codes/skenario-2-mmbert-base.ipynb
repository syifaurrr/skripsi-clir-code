```python
# # ==========================================
# # CELL 3: CLASS FAISS RETRIEVER
# # ==========================================
# import faiss
# import numpy as np
# from sentence_transformers import SentenceTransformer

# class FaissRetriever(pt.Transformer):
#     def __init__(self, model_name, index_path=None, doc_df=None, batch_size=32):
#         self.model = SentenceTransformer(model_name)
#         self.index = None
#         self.doc_map = {} 
#         self.batch_size = batch_size
        
#         # Jika ada dokumen, langsung buat index
#         if doc_df is not None:
#             self._index_docs(doc_df)
            
#     def _index_docs(self, doc_df):
#         print(f"Encoding {len(doc_df)} dokumen ke vektor (bisa agak lama)...")
#         docs = doc_df['text_clean'].tolist()
#         docnos = doc_df['docno'].tolist()
        
#         # Encode (Doc Arab) -> Vector
#         embeddings = self.model.encode(docs, batch_size=self.batch_size, show_progress_bar=True, convert_to_numpy=True)
#         faiss.normalize_L2(embeddings) # Penting untuk Cosine Similarity
        
#         # Build Index
#         dim = embeddings.shape[1]
#         self.index = faiss.IndexFlatIP(dim) # Inner Product (~Cosine Sim)
#         self.index.add(embeddings)
        
#         # Mapping ID FAISS -> Docno Asli
#         self.doc_map = {i: docno for i, docno in enumerate(docnos)}
#         print("Indexing selesai.")

#     def transform(self, queries):
#         # Encode (Query Indo) -> Vector
#         # Model Multilingual otomatis memetakan Indo & Arab ke ruang yang sama
#         q_texts = queries['query'].tolist()
#         q_embeddings = self.model.encode(q_texts, batch_size=self.batch_size, show_progress_bar=False, convert_to_numpy=True)
#         faiss.normalize_L2(q_embeddings)
        
#         # Search Top-1000
#         k = 1000
#         scores, ids = self.index.search(q_embeddings, k)
        
#         results = []
#         for i, qid in enumerate(queries['qid']):
#             for j in range(k):
#                 idx = ids[i][j]
#                 if idx != -1:
#                     results.append({
#                         'qid': str(qid),
#                         'docno': self.doc_map[idx],
#                         'score': scores[i][j],
#                         'rank': j + 1
#                     })
#         return pd.DataFrame(results)
# ==========================================
# CELL 3: CLASS FAISS RETRIEVER (PATCHED)
# ==========================================
import faiss
import numpy as np
import pandas as pd
import pyterrier as pt
from sentence_transformers import SentenceTransformer

class FaissRetriever(pt.Transformer):
    def __init__(self, model_name, index_path=None, doc_df=None, batch_size=32):
        self.model = SentenceTransformer(model_name)
        
        # --- PATCH KHUSUS DISTILBERT ---
        # Memaksa tokenizer untuk TIDAK mengeluarkan 'token_type_ids'
        if hasattr(self.model, 'tokenizer') and 'token_type_ids' in self.model.tokenizer.model_input_names:
            self.model.tokenizer.model_input_names.remove('token_type_ids')
        # -------------------------------
            
        self.index = None
        self.doc_map = {} 
        self.batch_size = batch_size
        
        # Jika ada dokumen, langsung buat index
        if doc_df is not None:
            self._index_docs(doc_df)
            
    def _index_docs(self, doc_df):
        print(f"Encoding {len(doc_df)} dokumen ke vektor (bisa agak lama)...")
        docs = doc_df['text_clean'].tolist()
        docnos = doc_df['docno'].tolist()
        
        # Encode (Doc Arab) -> Vector
        embeddings = self.model.encode(docs, batch_size=self.batch_size, show_progress_bar=True, convert_to_numpy=True)
        faiss.normalize_L2(embeddings) # Penting untuk Cosine Similarity
        
        # Build Index
        dim = embeddings.shape[1]
        self.index = faiss.IndexFlatIP(dim) # Inner Product (~Cosine Sim)
        self.index.add(embeddings)
        
        # Mapping ID FAISS -> Docno Asli
        self.doc_map = {i: docno for i, docno in enumerate(docnos)}
        print("Indexing selesai.")

    def transform(self, queries):
        # Encode (Query Indo) -> Vector
        q_texts = queries['query'].tolist()
        q_embeddings = self.model.encode(q_texts, batch_size=self.batch_size, show_progress_bar=False, convert_to_numpy=True)
        faiss.normalize_L2(q_embeddings)
        
        # Search Top-1000
        k = 1000
        scores, ids = self.index.search(q_embeddings, k)
        
        results = []
        for i, qid in enumerate(queries['qid']):
            for j in range(k):
                idx = ids[i][j]
                if idx != -1:
                    results.append({
                        'qid': str(qid),
                        'docno': self.doc_map[idx],
                        'score': scores[i][j],
                        'rank': j + 1
                    })
        return pd.DataFrame(results)
```

```python
# ==========================================
# CELL 4: EKSEKUSI SKENARIO 2 (FINAL & BERSIH)
# ==========================================
import pandas as pd
import os
import pyterrier as pt
import re
import pyarabic.araby as araby

# --- 1. Definisi Fungsi Normalisasi Lokal ---
def normalize_arabic(text):
    if not isinstance(text, str): return ""
    text = araby.strip_tashkeel(text)
    text = araby.strip_tatweel(text)
    text = araby.normalize_alef(text)
    text = re.sub(r'[ى]', 'ي', text) 
    text = re.sub(r'ؤ', 'و', text)
    text = re.sub(r'ئ', 'ي', text)
    return text

# --- 2. Konfigurasi Path ---
DATA_PATH = './skripsi-clir-code/data'
raw_docs_path = os.path.join(DATA_PATH, 'raw/fathul_muin.csv')
queries_indo_path = os.path.join(DATA_PATH, 'queries/queries_indo.csv')
qrels_path = os.path.join(DATA_PATH, 'queries/qrels.csv')

# --- 3. Load & Preprocess Dokumen ---
print("Loading data...")
df_docs = pd.read_csv(raw_docs_path)

# Asumsi kolom teks Anda bernama 'text' atau 'text_arabic'
text_col = 'text' if 'text' in df_docs.columns else 'text_arabic'

# Pembersihan ID (Strip spasi dan hapus .0 jika ada)
df_docs['docno'] = df_docs['docno'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)
df_docs['text_clean'] = df_docs[text_col].astype(str).apply(normalize_arabic)

# --- 4. Load Queries ---
df_queries = pd.read_csv(queries_indo_path)
df_queries['qid'] = df_queries['qid'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)

# --- 5. Load Qrels ---
qrels = pd.read_csv(qrels_path)
qrels['qid'] = qrels['qid'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)
qrels['docno'] = qrels['docno'].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)
qrels['label'] = qrels['label'].astype(int)

# Cek sekilas format ID
print(f"Format sampel df_queries QID : {df_queries['qid'].iloc[0]!r}")
print(f"Format sampel qrels QID      : {qrels['qid'].iloc[0]!r}")
print(f"Format sampel df_docs DOCNO  : {df_docs['docno'].iloc[0]!r}")
print(f"Format sampel qrels DOCNO    : {qrels['docno'].iloc[0]!r}")

# --- 6. Indexing & Retrieval ---
MODEL_NAME = 'jhu-clsp/mmBERT-base' 
print(f"\nMembangun Index Dense menggunakan model: {MODEL_NAME}...")

# Pastikan class FaissRetriever sudah di-run di Cell 3
dense_retriever = FaissRetriever(
    model_name=MODEL_NAME,
    doc_df=df_docs[['docno', 'text_clean']],
    batch_size=32
)

# --- 7. Evaluasi PyTerrier ---
print("\nMenjalankan Eksperimen Skenario 2...")
pipeline = pt.Experiment(
    [dense_retriever],
    df_queries,
    qrels,
    eval_metrics=["map", "recip_rank", "recall_10", "recall_20"],
    names=[f"Dense ({MODEL_NAME})"],
    validate="ignore"  # Mengabaikan warning merah PyTerrier
)

print("\n=== HASIL EVALUASI SKENARIO 2 ===")
print(pipeline)

# Simpan hasil
pipeline.to_csv('hasil_skenario_2.csv')
print("\nHasil tersimpan di 'hasil_skenario_2.csv'")
```

```python
# ==========================================
# CELL 5: KOMPARASI BASELINE VS FINE-TUNED + ANALISIS DETAIL
# ==========================================
import pyterrier as pt
import pandas as pd
from IPython.display import display

print("1. Memuat Model 1: Baseline (Belum Fine-Tuning)...")
BASE_MODEL_NAME = 'jhu-clsp/mmBERT-base'
retriever_base = FaissRetriever(
    model_name=BASE_MODEL_NAME, 
    doc_df=df_docs[['docno', 'text_clean']], 
    batch_size=32
)

print("\n2. Memuat Model 2: Fine-Tuned (Hasil JH-POLO)...")
# [!] PENTING: Ubah path ini jika Anda menggunakan "Add Data" dari notebook lain
# FINETUNED_MODEL_PATH = '/kaggle/input/datasets/syifaur/mmbert-base-bs8-ep6-lr3e-05-wr0-1-seq128' 
# NORMALISASI
FINETUNED_MODEL_PATH = '/kaggle/input/datasets/syifaur/mmbert-base-bs8-ep4-lr3e-05-wr0-1-seq128-norm'

retriever_finetuned = FaissRetriever(
    model_name=FINETUNED_MODEL_PATH, 
    doc_df=df_docs[['docno', 'text_clean']], 
    batch_size=32
)

print("\n3. Menjalankan PyTerrier Experiment (perquery=True)...")
eval_metrics_list = ["recip_rank", "success_10", "success_20", "success_50", "success_100"]

# Jalankan eksperimen per-kueri
hasil_eval_perq = pt.Experiment(
    [retriever_base, retriever_finetuned], # Masukkan kedua model
    df_queries,
    qrels,
    eval_metrics=eval_metrics_list,
    names=["Baseline (mmBERT-base)", "mmBERT-base Fine-Tuned (JH-POLO)"],
    validate="ignore",
    perquery=True # <-- KUNCI ANALISIS DETAIL
)

# Format Ulang Tabel (Pivot) agar metrik menjadi kolom (karena perquery=True membuat tabel jadi format panjang)
hasil_pivot = hasil_eval_perq.pivot_table(
    index=['name', 'qid'],
    columns='measure',
    values='value'
).reset_index()

# Pastikan tipe data qid sama (string)
hasil_pivot['qid'] = hasil_pivot['qid'].astype(str)
df_queries['qid'] = df_queries['qid'].astype(str)

# Gabungkan dengan metadata dari df_queries (Berisi query_type dan teks pertanyaan bahasa Indonesianya)
hasil_dengan_tipe = pd.merge(
    hasil_pivot,
    df_queries[['qid', 'query_type', 'query']],
    on='qid',
    how='left'
)

# ==========================================
# 4. MENAMPILKAN HASIL KESELURUHAN (OVERALL)
# ==========================================
print("\n" + "="*60)
print("🎯 HASIL KESELURUHAN (OVERALL)")
print("="*60)

overall_metrics = hasil_dengan_tipe.groupby('name')[eval_metrics_list].mean().reset_index()

for k in [10, 20, 50, 100]:
    overall_metrics[f'success_{k} (%)'] = (overall_metrics[f'success_{k}'] * 100).round(2)
overall_metrics['recip_rank'] = overall_metrics['recip_rank'].round(4)

display(overall_metrics[['name', 'recip_rank', 'success_10 (%)', 'success_20 (%)', 'success_50 (%)', 'success_100 (%)']])

# ==========================================
# 5. MENAMPILKAN HASIL BERDASARKAN TIPE KUERI
# ==========================================
print("\n" + "="*60)
print("📊 HASIL BERDASARKAN TIPE KUERI")
print("="*60)

per_type_metrics = hasil_dengan_tipe.groupby(['name', 'query_type'])[eval_metrics_list].mean().reset_index()

for k in [10, 20, 50, 100]:
    per_type_metrics[f'success_{k} (%)'] = (per_type_metrics[f'success_{k}'] * 100).round(2)
per_type_metrics['recip_rank'] = per_type_metrics['recip_rank'].round(4)

display(per_type_metrics[['name', 'query_type', 'recip_rank', 'success_10 (%)', 'success_20 (%)', 'success_50 (%)', 'success_100 (%)']])

# ==========================================
# 6. MENAMPILKAN DETAIL UNTUK SETIAP KUERI (HIT/MISS)
# ==========================================
print("\n" + "="*60)
print("🔍 ANALISIS DETAIL PER-KUERI")
print("="*60)

detail_kueri = hasil_dengan_tipe.copy()

# Konversi Reciprocal Rank kembali menjadi Rank
detail_kueri['rank_ditemukan'] = detail_kueri['recip_rank'].apply(
    lambda x: int(round(1/x)) if x > 0 else "Tidak Ketemu"
)

# Label visual Hit / Miss
for k in [10, 20, 50, 100]:
    detail_kueri[f'Hit@{k}'] = detail_kueri[f'success_{k}'].apply(
        lambda x: '✅ Hit' if x == 1.0 else '❌ Miss'
    )

# Urutkan berdasarkan Model -> Tipe Kueri -> Nomor Kueri
# qid diubah ke integer sementara agar pengurutannya 1,2,3.. (bukan 1, 10, 11..)
detail_kueri['qid_int'] = pd.to_numeric(detail_kueri['qid'], errors='coerce').fillna(0).astype(int)
detail_kueri = detail_kueri.sort_values(by=['name', 'query_type', 'qid_int'])

kolom_tampil = ['name', 'qid', 'query_type', 'query', 'rank_ditemukan', 'Hit@10', 'Hit@20', 'Hit@50', 'Hit@100']
tabel_detail_final = detail_kueri[kolom_tampil]

display(tabel_detail_final.head(15)) # Tampilkan sampel 15 baris pertama

# Simpan semua tabel ke CSV untuk referensi skripsi
tabel_detail_final.to_csv('skenario2_mmbert-base_detail_per_kueri.csv', index=False)
per_type_metrics.to_csv('skenario2_mmbert-base_evaluasi_tipe_kueri.csv', index=False)
overall_metrics.to_csv('skenario2_mmbert-base_evaluasi_overall.csv', index=False)

print("\n📑 Seluruh laporan berhasil disimpan dalam format CSV!")
```
